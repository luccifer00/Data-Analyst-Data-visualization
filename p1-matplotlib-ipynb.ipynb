# Ejercicios de matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## 1. Gr√°fica una recta de (-7,-7) a (7,7) de color rosa

```python
x = np.linspace(-7, 7, 20)
y=x
plt.figure(figsize=(8,8))
plt.plot(x,y, color='pink')
plt.show()
```

## 2. Obt√©n el siguiente resultado utilizando subplot y la ecuaci√≥n anterior

```python
plt.subplot(2,2,1)
plt.plot(x,y, color='pink')
plt.subplot(2,2,2)
plt.plot(x,y,color='blue')
plt.subplot(2,2,3)
plt.plot(x,y, color='red')
plt.subplot(2,2,4)
plt.plot(x,y,color='magenta')
plt.show()
```

## 3. Obt√©n la siguiente gr√°fica de barras 

```python
eti= ["a","b","c","d"]
valores=[10,20,30,40]
col=['blue','yellow', 'cyan', 'magenta']
#plt.figure(figsize=(10,8))
plt.bar(eti, valores, color=col)
plt.show()
```

## 4. Obtener la siguiente gr√°fica

```python
val2=[i+10 for i in valores]
fig, ax = plt.subplots(figsize=(8,6))
x= np.arange(len(eti))
anch=0.5
plt.bar(x - 0.25, height=valores, width=anch, color='cyan')
plt.bar(x + 0.25, height=val2, width=anch, color='magenta')

ax.set_xticks(x)
ax.set_xticklabels(eti)
plt.show()
```

## 7. Guarda el dataset de titanic en una variable

```python
tit=sns.load_dataset("titanic")
tit.head()
```

## 8. Muestra en una grafica de barras el n√∫mero de hombres y mujeres que sobrevivieron

```python
onlysurvived= tit[tit['survived']==1]
plt.bar(onlysurvived.iloc[:,2].unique(),onlysurvived.iloc[:,2].value_counts(), width=0.8, color='green')
plt.show()
```

```python
onlysurvived= tit[tit['survived']==1]
sns.countplot(x='sex', data=onlysurvived, color='green')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

## 9. Muestra en una grafica de barras agrupada la cantidad de hombres y mujeres que sobrevivieron y la cantidad de los que no sobrevivieron

```python
cus={1:'blue', 0:'red'}
sns.countplot(x='sex', hue='survived', data=tit, palette=cus, order=['female','male'], hue_order=[1,0])
plt.legend().remove
plt.title('Distribucion de hombres y mujeres que sobrevivieron')
plt.xlabel('Genero')
plt.ylabel('Cantidad')
plt.legend().remove()
plt.show()
```

## 10. Muestra en una grafica de barras agrupada la cantidad de hombres y mujeres de cada clase (Pclass)

```python
colo={1:'blue',2:'red',3:'green'}
sns.countplot(x='sex', hue='pclass', data=tit, palette=colo)
plt.show()
```

## 12. Muestra en una gr√°fica lineal las edades de los pasajeros que sobrevivieron y en otra la de los pasajeros que no sobrevivieron

```python
colife={1:'blue',0:'red'}
survi=tit[tit.survived==1]
nosurvi=tit[tit.survived==0]
plt.figure(figsize=(10,2))
plt.plot(survi.iloc[:,3])
plt.plot(nosurvi.iloc[:,3], color='red')
plt.grid(False)
plt.show()
```

```python
# Filtrar datos de pasajeros que sobrevivieron y que no sobrevivieron
survived_data = tit[tit['survived'] == 1]
not_survived_data = tit[tit['survived'] == 0]

# Crear el gr√°fico con Seaborn
plt.figure(figsize=(10, 3))
sns.lineplot(data=survived_data, x=survived_data.index, y='age', color='blue', label='Sobrevivieron')
sns.lineplot(data=not_survived_data, x=not_survived_data.index, y='age', color='red', label='No sobrevivieron')
plt.title('Edades de los pasajeros que sobrevivieron y no sobrevivieron')
plt.xlabel('√çndice del pasajero')
plt.ylabel('Edad')
plt.legend()
plt.show()

```

## 13. Muestra en una grafica de caja la edad de los que sobrevivieron y de los que no sobrevivieron


Nota: Para poder graficarlo hay que reemplazar los NaN por el promedio de edad. 

```python
alldata=[not_survived_data['age'].fillna(tit['age'].mean()), survived_data['age'].fillna(tit['age'].mean())]
plt.figure(figsize=(8,6))
plt.boxplot(alldata, labels=['No Supervivientes','Supervivientes'])
plt.grid(False)
plt.show()
```

## 14. Mostrar en una gr√°fica de caja por clase la edad de los pasajeros

```python
datis= [tit[tit['pclass']==i]['age'].fillna(tit['age'].mean()) for i in sorted(tit['pclass'].unique())]
class_labels = {1: "Clase 1", 2: "Clase 2", 3: "Clase 3"}

#fig, ax = plt.subplots()
plt.boxplot(datis, labels=[class_labels[i] for i in sorted(tit['pclass'].unique())])
plt.show()
```

## 17. Muestra en un histograma la edad de los pasajeros que sobrevivieron y la edad de los que no

```python
survi=tit[tit.survived ==1]
nosurvi=tit[tit.survived==0]
plt.hist(survi.iloc[:,3], bins=20, alpha=0.2,color='blue', label='Supervivientes')
plt.hist(nosurvi.iloc[:,3], bins=20, alpha=0.2,color='red', label='No Supervivientes')
plt.legend()
plt.show()
```

## 18. Muestra en un histograma el costo de boleto por clase

```python
prim=tit[tit.pclass==1]
seco=tit[tit.pclass==2]
ter=tit[tit.pclass==3]
plt.hist(prim['fare'], bins=10, color='blue',alpha=0.2, label='Primera')
plt.hist(seco['fare'], bins=10, color='red',alpha=0.5, label='Segunda')
plt.hist(ter['fare'], bins=10, color='yellow',alpha=0.2, label='Tercera')
plt.legend()
plt.xlabel('Precio')
plt.ylabel('Cantidad de Personas')
plt.title('Hist de precio del ticket por clase')
plt.show()
```
