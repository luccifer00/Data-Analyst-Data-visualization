## 1.	Graficar por municipio el número de habitaciones de cada tipo

```python

df.room_type.value_counts()
munis = df['municipio'].unique().tolist()
res=[]
for mu in munis:
    temp = df[df['municipio']== mu]['room_type'].value_counts()
    res.append([mu, temp])
print(res[0][1].values)
plt.figure(figsize=(15, 8))  # Adjust figsize as needed
for i, mu in enumerate(munis):
    room_types = res[i][1].index
    counts = res[i][1].values
    plt.bar([f"{mu} - {room_type}" for room_type in room_types], counts, label=mu)

plt.xlabel('Room Type')
plt.ylabel('Count')
plt.title('Room Type Distribution by Municipality')
plt.xticks(rotation=45, ha='right')  # Rotates x labels for better readability
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have already imported pandas and have your DataFrame stored in 'df'

# Your existing code for counting room types and municipalities
df.room_type.value_counts()
munis = df['municipio'].unique().tolist()
res=[]
for mu in munis:
    temp = df[df['municipio'] == mu]['room_type'].value_counts()
    res.append([mu, temp])

# Plotting
plt.figure(figsize=(15, 8))  # Adjust figsize as needed
room_types = df['room_type'].unique()
bar_width = 0.2
num_bars = len(munis)
for i, room_type in enumerate(room_types):
    x = range(num_bars)
    heights = [res[j][1].get(room_type, 0) for j in range(num_bars)]
    plt.bar([pos + i * bar_width for pos in x], heights, bar_width, label=room_type)

plt.xlabel('Municipality')
plt.ylabel('Count')
plt.title('Room Type Distribution by Municipality')
plt.xticks([r + bar_width * (len(room_types) - 1) / 2 for r in range(num_bars)], munis)
plt.legend()
plt.tight_layout()
plt.show()
```
